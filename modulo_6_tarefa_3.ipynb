```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns

df = pd.read_csv(r'SINASC_RO_2019.csv', sep = ',')
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df.loc[(df['APGAR5'] >= 0) & (df['APGAR5'] <= 3), 'APGAR5_QUEBRA'] = 'asfixia grave'
df.loc[(df['APGAR5'] >= 4) & (df['APGAR5'] <= 5), 'APGAR5_QUEBRA'] = 'asfixia moderada'
df.loc[(df['APGAR5'] >= 6) & (df['APGAR5'] <= 7), 'APGAR5_QUEBRA'] = 'asfixia leve'
df.loc[(df['APGAR5'] >= 8) & (df['APGAR5'] <= 10), 'APGAR5_QUEBRA'] = 'normal'
```

```python
df['APGAR5'].value_counts()
```

```python
df['APGAR5_QUEBRA'].value_counts()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
plt.figure(figsize = (10, 6))

ax1 = sns.barplot(x = 'SEXO',
                 y = 'APGAR5',
                 hue = 'APGAR5_QUEBRA',
                 data = df,
                 ci = None)
plt.legend(loc = 'upper right')
```

```python
ax = sns.catplot(x = 'SEXO',
            y = 'APGAR5',
            hue = 'APGAR5_QUEBRA',
            col = 'GESTACAO',
            kind = 'bar',
            data = df,
            ci = None)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
df_tabela_cruzada = pd.crosstab(df['APGAR5_QUEBRA'], df['GESTACAO'])
```

```python
df_tabela_cruzada
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
df_tabela_cruzada.sum(axis = 1)
```

```python
freq_tabela_cruzada = df_tabela_cruzada.div(df_tabela_cruzada.sum(axis = 1), axis = 0)
round(freq_tabela_cruzada * 100, 2)
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
freq_tabela_cruzada.plot.bar(figsize = (10, 5))

plt.legend(loc = 'upper left')
plt.xlabel('Quadro asfixia')
```
